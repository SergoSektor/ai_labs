## 2. Описание исходного процесса: Подбор музыкального жанра

В рамках данной работы разработан технологический процесс подбора музыкального жанра для пользователя на основе его предпочтений и текущей ситуации. Этот процесс является упрощенной моделью функционирования интеллектуальной системы рекомендаций.

### 2.1. Формальное описание технологического процесса

Технологический процесс состоит из четырех основных этапов:

1.  **Сбор предпочтений пользователя (User Preference Collection):** На этом этапе система собирает информацию о желаниях пользователя. Это могут быть как явные запросы (например, "хочу что-то энергичное"), так и неявные данные (история прослушиваний, текущее настроение).
2.  **Фильтрация по основным критериям (Primary Filtering):** Полученные предпочтения используются для первичного отбора жанров, которые потенциально подходят пользователю.
3.  **Оценка жанров по дополнительным параметрам (Secondary Evaluation):** Оставшиеся жанры оцениваются с учетом дополнительных факторов, таких как время суток, день недели, погодные условия, что позволяет уточнить рекомендацию.
4.  **Выбор оптимального жанра (Optimal Genre Selection):** На основе всей собранной и обработанной информации система предлагает один или несколько наиболее подходящих музыкальных жанров, учитывая при этом возможные ограничения.

### 2.2. Обоснование выбора операций и их характеристик

Выбор данных операций обусловлен необходимостью демонстрации как вероятностных, так и нечетких характеристик в рамках гибридной экспертной системы. Каждая операция включает в себя элементы неопределенности, что делает ее подходящей для применения различных методов интеллектуального анализа.

**Детализация операций и их характеристик:**

**1. Сбор предпочтений пользователя (User Preference Collection)**
*   **Вход:** Явные запросы пользователя (текстовые, голосовые), история прослушиваний, данные о текущем настроении (если система способна его определять).
*   **Выход:** Набор параметров, характеризующих желаемую музыку: `энергия` (числовая шкала), `темп` (числовая шкала), `настроение` (нечеткое множество: `грустное`, `нейтральное`, `радостное`), `наличие вокала` (вероятностное: `да`/`нет`), `инструменты` (категориальные).
*   **Характеристики:**
    *   **Вероятностные:**
        *   Вероятность предпочтения энергичной музыки (на основе истории): $P(Энергичная\_музыка)$.
    *   **Нечеткие:**
        *   `Настроение`: Определяется по степени принадлежности к нечетким множествам (`грустное`, `нейтральное`, `радостное`). Например, настроение может быть на 70% радостным и на 30% нейтральным.
        *   `Желаемая энергия`: `Низкая`, `Средняя`, `Высокая` (определяется как нечеткие множества).

**2. Фильтрация по основным критериям (Primary Filtering)**
*   **Вход:** Параметры предпочтений пользователя.
*   **Выход:** Предварительный список подходящих музыкальных жанров.
*   **Характеристики:**
    *   **Вероятностные:**
        *   Вероятность совпадения жанра с общими предпочтениями пользователя: $P(Жанр\_соответствует | Предпочтения)$.
    *   **Нечеткие:**
        *   Степень соответствия жанра желаемому настроению и энергии (на основе нечеткой логики).

**3. Оценка жанров по дополнительным параметрам (Secondary Evaluation)**
*   **Вход:** Предварительный список жанров, дополнительные контекстные данные: `время суток` (утро, день, вечер, ночь), `день недели` (будний, выходной), `погодные условия` (солнечно, пасмурно, дождь).
*   **Выход:** Отранжированный список жанров.
*   **Характеристики:**
    *   **Вероятностные:**
        *   Вероятность того, что жанр подходит для текущего времени суток: $P(Жанр\_подходит | Время\_суток)$.
        *   Вероятность того, что жанр подходит для дня недели: $P(Жанр\_подходит | День\_недели)$.
    *   **Нечеткие:**
        *   Степень соответствия жанра погодным условиям: `Приятно\_для\_солнечной\_погоды`, `Уютно\_для\_дождливой\_погоды`.\n

**4. Выбор оптимального жанра (Optimal Genre Selection)**
*   **Вход:** Отранжированный список жанров, пользовательские ограничения (например, `максимальное количество предложений`, `исключение недавно проигранных жанров`).
*   **Выход:** Итоговая рекомендация – один или несколько музыкальных жанров.
*   **Характеристики:**
    *   Оптимизация на основе взвешенных критериев (вероятностных и нечетких), учитывающая ограничения. Например, минимизация "дискомфорта" пользователя при выборе жанра или максимизация "удовольствия".


## 3. Построение функциональной схемы

Ниже представлено текстовое описание функциональной схемы процесса подбора музыкального жанра.

**Этап 1: Сбор предпочтений пользователя.**
*   **Вход:** Запросы пользователя, история прослушиваний, данные о настроении.
*   **Выход:** Параметры желаемой музыки (энергия, темп, настроение, вокал, инструменты) с вероятностными и нечеткими характеристиками.
*   **Альтернативы:** Только явные запросы или комбинация явных запросов и истории прослушиваний.

**Этап 2: Фильтрация по основным критериям.**
*   **Вход:** Параметры предпочтений пользователя.
*   **Выход:** Предварительный список подходящих музыкальных жанров.
*   **Альтернативы:** Строгая пороговая фильтрация или нечеткая фильтрация с использованием α-срезов.

**Этап 3: Оценка жанров по дополнительным параметрам.**
*   **Вход:** Предварительный список жанров, контекстные данные (время суток, день недели, погода).
*   **Выход:** Отранжированный список жанров.
*   **Альтернативы:** Учет всех доступных дополнительных параметров или только наиболее значимых.

**Этап 4: Выбор оптимального жанра.**
*   **Вход:** Отранжированный список жанров, пользовательские ограничения.
*   **Выход:** Итоговая рекомендация – один или несколько музыкальных жанров.

**Обоснование соответствия области Парето:**

Область Парето здесь представляет собой набор жанров, для которых невозможно улучшить соответствие одному критерию (например, "энергия") без ухудшения соответствия другому (например, "настроение"). Оптимизация направлена на поиск жанров на границе Парето, чтобы предложить пользователю наилучший компромисс между противоречивыми предпочтениями, либо несколько Парето-оптимальных вариантов.

### 4.1. Решение оптимизационных задач с вероятностными характеристиками


```python
import numpy as np

# Пример определения жанров с вероятностными характеристиками
genres_data_prob = {
    "Pop": {"energy": ("high", 0.7), "vocal": ("yes", 0.9), "mood": ("happy", 0.8), "tempo": ("fast", 0.6)},
    "Classical": {"energy": ("low", 0.2), "vocal": ("no", 0.8), "mood": ("calm", 0.9), "tempo": ("slow", 0.7)},
    "Rock": {"energy": ("high", 0.8), "vocal": ("yes", 0.8), "mood": ("energetic", 0.7), "tempo": ("fast", 0.7)},
    "Jazz": {"energy": ("medium", 0.5), "vocal": ("optional", 0.5), "mood": ("calm", 0.7), "tempo": ("medium", 0.6)},
    "Electronic": {"energy": ("very high", 0.9), "vocal": ("optional", 0.6), "mood": ("energetic", 0.8), "tempo": ("very fast", 0.8)},
}

class GenreClassifierProbabilistic:
    def __init__(self, genres_data):
        self.genres_data = genres_data

    def get_genre_probability(self, genre, feature, value):
        feature_data = self.genres_data.get(genre, {}).get(feature)
        if feature_data and feature_data[0] == value:
            return feature_data[1]
        return 0.1 # Низкая вероятность, если нет прямого совпадения

    def recommend_genre_probabilistic(self, user_preferences):
        recommendations = {}
        for genre, data in self.genres_data.items():
            total_prob = 1.0
            for feature, desired_value in user_preferences.items():
                total_prob *= self.get_genre_probability(genre, feature, desired_value)
            recommendations[genre] = total_prob
        
        sorted_reсommendations = sorted(recommendations.items(), key=lambda item: item[1], reverse=True)
        return sorted_reсommendations

# Пример использования:
user_prefs = {"energy": "high", "vocal": "yes", "mood": "happy"}
classifier = GenreClassifierProbabilistic(genres_data_prob)
recs = classifier.recommend_genre_probabilistic(user_prefs)
print("Вероятностные рекомендации:", recs)


# Дополнительный пример: моделирование задержек в процессе обработки запроса
class RequestProcessor:
    def __init__(self, processing_time_mean, processing_time_std):
        self.processing_time_mean = processing_time_mean
        self.processing_time_std = processing_time_std

    def process_request(self):
        return np.random.normal(self.processing_time_mean, self.processing_time_std)

# Пример создания обработчиков для разных этапов
preference_collector = RequestProcessor(processing_time_mean=2, processing_time_std=0.5)
filter_processor = RequestProcessor(processing_time_mean=1, processing_time_std=0.2)
evaluation_processor = RequestProcessor(processing_time_mean=3, processing_time_std=1)

# Симуляция общего времени обработки запроса
def simulate_total_time(num_simulations=1000):
    total_times = []
    for _ in range(num_simulations):
        time = preference_collector.process_request() + filter_processor.process_request() + evaluation_processor.process_request()
        total_times.append(time)
    return np.mean(total_times), np.std(total_times)

mean_time, std_time = simulate_total_time()
print(f"Среднее время обработки запроса: {mean_time:.2f} \u00b1 {std_time:.2f} секунд")
```

С увеличением `min_energy` система отдает предпочтение более "энергичным" жанрам (например, "metal" и "rock"), исключая менее энергичные. Это демонстрирует, как изменение одного параметра может кардинально изменить оптимальное решение.
*   **Чувствительность к `max_recommendations`:** Изменение максимального количества рекомендаций напрямую влияет на разнообразие предложений. Увеличение этого параметра позволяет включить больше подходящих жанров, но при этом может снизить "точность" каждой отдельной рекомендации, так как в список попадают жанры с более низкой вероятностью соответствия.

**Выводы об эффективности стратегий:**
Гибридная экспертная система, комбинирующая вероятностные и нечеткие подходы, обладает высокой адаптивностью. Вероятностный подход хорошо справляется с количественной оценкой соответствия, а нечеткая логика позволяет эффективно работать с качественными и субъективными предпочтениями. Анализ чувствительности к ограничениям критически важен для понимания того, как внешние факторы влияют на принимаемые решения, позволяя проектировать более устойчивые и гибкие системы.
